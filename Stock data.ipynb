```python

import yfinance as yf

import matplotlib.pyplot as plt


gme_data = yf.Ticker("GME").history(period="max")


def make_graph(stock_data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data.index, stock_data['Close'], label='Closing Price')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Stock Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.show()


make_graph(gme_data, "GameStop (GME) Stock Price Over Time")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

tesla_data = yf.Ticker("TSLA").history(period="max")

def make_graph(stock_data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data.index, stock_data['Close'], label='Closing Price')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Stock Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.show()

make_graph(tesla_data, "Tesla (TSLA) Stock Price Over Time")
```
